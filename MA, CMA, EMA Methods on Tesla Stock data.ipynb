```python
import pandas_datareader as pdr
import pandas as pd
from datetime import datetime
```

```python
df_tesla=pdr.data.DataReader('TSLA', data_source='stooq', start='2017-02-10', end='2022-02-09')
```

```python
df_tesla.head()
```

```python
df_tesla['Open'].plot(figsize=(12,6))
```

```python
df_tesla['Open: 10days rolling']=df_tesla['Open'].rolling(window=10, min_periods=1).mean()
```

```python
df_tesla[['Open', 'Open: 10days rolling']].plot(xlim = ['2020-01-01', '2021-01-01'], figsize=(10,6))
```

```python
df_tesla['Open: 30days rolling']=df_tesla['Open'].rolling(window=30, min_periods=1).mean()
df_tesla['Open: 50days rolling']=df_tesla['Open'].rolling(window=50, min_periods=1).mean()
```

```python
df_tesla[['Open', 'Open: 10days rolling', 'Open: 30days rolling','Open: 50days rolling' ]].plot(xlim = ['2020-01-01', '2021-01-01'], figsize=(10,6))
```

```python
# Cummulative moving average implementation (CMA)
# Expanding function
df_tesla['Open'].expanding().mean().plot(figsize=(15,6))
```

```python
# Exponential moving average
# Smoothing factor = 0.1
df_tesla['EMA_0.1']=df_tesla['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1' ]].plot(xlim = ['2020-01-01', '2021-01-01'], figsize=(10,6))
```

```python
# Exponential Weight moving average
df_tesla['EMA_5days'] = df_tesla['Open'].ewm(span=5).mean()
```

```python
df_tesla[['Open', 'EMA_5days']].plot(figsize=(15,6))
```

### Conclusion: Out of all the averages, Exponentiol Weighted moving average is best.
